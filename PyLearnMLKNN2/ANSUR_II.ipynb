```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
female_data = pd.read_csv('input/ANSUR_II_FEMALE_Public.csv')

female_data.head()
```

```python
male_data = pd.read_csv('input/ANSUR_II_MALE_Public.csv', encoding='latin-1')

male_data
```

```python
 data = pd.concat([female_data, male_data])
 data
```

```python
data['weightkg'] = data['weightkg'] / 10
data['stature'] = data['stature'] / 10
data['Gender'] = data['Gender'].replace(['Female', 'Male'], [0, 1])
```

```python
cmap = ['red', 'blue']
plt.scatter(female_data['stature']/10, female_data['weightkg']/10, c= 'red', alpha=0.2)
plt.scatter(male_data['stature']/10, male_data['weightkg']/10, c= 'blue', alpha=0.2)
# plt.scatter(data['stature'], data['weightkg'], c=data['Gender'], cmap=matplotlib.colors.ListedColormap(cmap), alpha=0.2)
plt.title('Display weight and stature data of women and men')
plt.xlabel('Stature (cm)')
plt.ylabel('Weight (Kg)')
plt.legend(['Female', 'Male'])
plt.show()
```

### Why is the data of men higher than the data of women?

Because men are taller.

### Why is the data of men more right than the data of women?

Because men are heavier.

## Split dataset to train and test datasets

```python
from sklearn.model_selection import train_test_split
```

```python
training_data, testing_data = train_test_split(data[['weightkg','stature','Gender']], test_size=0.2, random_state=25)
```

## Implement and fit your object oriented KNN algorithm on the train dataset and test dataset with different values of k = 3, 5, 7, ... 

```python
testing_data[['weightkg','stature']].to_numpy()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from knn import KNN

Y_test = testing_data['Gender'].to_numpy()
for k in [3, 5, 7, 10, 15]:
    knn = KNN(k)
    knn.fit(training_data[['weightkg','stature']].to_numpy(), training_data['Gender'].to_numpy())

    score = knn.evaluate(testing_data[['weightkg','stature']].to_numpy(), Y_test)
    print(f'Score in k = {k} is {score*100:.1f}%.')
```

| k      | Score  |
| :---   | :----: |
| 3      | 83.1%  |
| 5      | 84.7%  |
| 7      | 84.9%  |
| 10     | 85.2%  |
| 15     | 85.1%  |

## Fit the scikit-learn KNN algorithm on the train dataset

```python
from sklearn.neighbors import KNeighborsClassifier

Y_test = testing_data['Gender'].to_numpy()
for k in [3, 5, 7, 10, 15]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(training_data[['weightkg','stature']], training_data['Gender'])
    
    Y_pred = knn.predict(testing_data[['weightkg','stature']])
    score = np.sum(Y_pred == Y_test) / Y_pred.shape[0]
    print(f'Score in k = {k} is {score*100:.1f}%.')
```

| k      | Score  |
| :---   | :----: |
| 3      | 83.3%  |
| 5      | 84.5%  |
| 7      | 84.8%  |
| 10     | 85.2%  |
| 15     | 85.1%  |
